<h1>7. The column 'Height' is in inches with a quotation mark, Convert to float
with decimal points. <h1>

```python
def height_to_dec(height):
    try:
        height_str = str(height)
        if "'" in height_str:
            feet, inches = height_str.split("'")
            total_inches = int(feet) * 12 + int(inches)
            height_decimal = total_inches / 12
        else:
            height_decimal = float(height_str)
    except Exception as e:
        height_decimal = float('nan')
    return height_decimal

df['Height'] = df['Height'].apply(height_to_dec)
```
